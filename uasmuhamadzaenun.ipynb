```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi keanggotaan untuk suhu
def fuzzy_membership_suhu(x):
    dingin = max(0, min((25 - x) / 10, 1))
    normal = max(0, min((x - 20) / 5, (30 - x) / 5))
    panas = max(0, min((x - 25) / 5, 1))
    return {'dingin': dingin, 'normal': normal, 'panas': panas}

# Fungsi keanggotaan untuk kelembapan
def fuzzy_membership_kelembapan(x):
    rendah = max(0, min((50 - x) / 20, 1))
    sedang = max(0, min((x - 40) / 10, (60 - x) / 10))
    tinggi = max(0, min((x - 50) / 20, 1))
    return {'rendah': rendah, 'sedang': sedang, 'tinggi': tinggi}

# Visualisasi fungsi keanggotaan untuk suhu
def plot_fuzzy_suhu():
    x = np.linspace(15, 35, 100)  # Rentang suhu
    dingin = [fuzzy_membership_suhu(i)['dingin'] for i in x]
    normal = [fuzzy_membership_suhu(i)['normal'] for i in x]
    panas = [fuzzy_membership_suhu(i)['panas'] for i in x]

    plt.figure(figsize=(8, 6))
    plt.plot(x, dingin, label='Dingin', color='blue')
    plt.plot(x, normal, label='Normal', color='green')
    plt.plot(x, panas, label='Panas', color='red')
    plt.title("Fungsi Keanggotaan Fuzzy - Suhu")
    plt.xlabel("Suhu (C)")
    plt.ylabel("Derajat Keanggotaan")
    plt.legend()
    plt.grid()
    plt.show()

# Visualisasi fungsi keanggotaan untuk kelembapan
def plot_fuzzy_kelembapan():
    x = np.linspace(30, 70, 100)  # Rentang kelembapan
    rendah = [fuzzy_membership_kelembapan(i)['rendah'] for i in x]
    sedang = [fuzzy_membership_kelembapan(i)['sedang'] for i in x]
    tinggi = [fuzzy_membership_kelembapan(i)['tinggi'] for i in x]

    plt.figure(figsize=(8, 6))
    plt.plot(x, rendah, label='Rendah', color='purple')
    plt.plot(x, sedang, label='Sedang', color='orange')
    plt.plot(x, tinggi, label='Tinggi', color='brown')
    plt.title("Fungsi Keanggotaan Fuzzy - Kelembapan")
    plt.xlabel("Kelembapan (%)")
    plt.ylabel("Derajat Keanggotaan")
    plt.legend()
    plt.grid()
    plt.show()

def main():
    print("Menampilkan Grafik Fungsi Keanggotaan Fuzzy:")
    plot_fuzzy_suhu()
    plot_fuzzy_kelembapan()

if __name__ == "__main__":
    main()
```
